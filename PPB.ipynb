```python
import time
import pandas as pd
import matplotlib.pyplot as plt
from sdv.metadata import SingleTableMetadata
from sdv.single_table import CTGANSynthesizer, TVAESynthesizer

# Select only the necessary columns for both data subset and metadata
data_subset = mimic_data[['ADMISSION_TYPE', 'ETHNICITY', 'HOSPITAL_EXPIRE_FLAG', 'GENDER']].iloc[:500]

# Define discrete columns and set up metadata
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=data_subset)

# Mark the columns as categorical as per our subset
for column in ['ADMISSION_TYPE', 'ETHNICITY', 'HOSPITAL_EXPIRE_FLAG', 'GENDER']:
    metadata.update_column(column_name=column, sdtype='categorical')

# Function to measure training time for each synthesizer
def measure_time(synthesizer_class, data_subset):
    start_time = time.time()
    model = synthesizer_class(metadata)
    model.fit(data_subset)
    return time.time() - start_time

# Measure time for CTGANSynthesizer
ctgan_time = measure_time(CTGANSynthesizer, data_subset)

# Measure time for TVAESynthesizer
tvae_time = measure_time(TVAESynthesizer, data_subset)

# Plot the results
labels = ['CTGAN', 'TVAE']
times = [ctgan_time, tvae_time]

plt.bar(labels, times, color=['blue', 'orange'])
plt.ylabel("Training Time (seconds)")
plt.title("CTGAN vs TVAE Training Time Comparison (500 rows, 5 columns)")
plt.show()
```
